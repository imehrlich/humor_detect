2. Create a dictionary of the CMU dataset

```python
import re 
pronunciations = dict()
file = 'cmudict-0.7b'
for line in open(file,encoding = 'latin-1').readlines():
    line = line.strip()
    if line.startswith(';'): continue
    word,phones = line.split("  ")
    word = word.rstrip("(0123)").lower()
    if word not in pronunciations:
        pronunciations[word] = []
    pronunciations[word].append(phones)
```

3. Count the number of alliteration in a sentence

```python
stopwords = nltk.corpus.stopwords.words("english")
no_punct = lambda x: re.sub(r'[^\w\s]', '', x)
import json
def removeNonWords(phrase):
    bad_chars = [';', ':', '!', '*', ',', '.', '-', '"', '?']
    for bad in bad_chars:
        if (bad in phrase):
            phrase = phrase.replace(bad, '')
    line = phrase.split()
    return line
    
def alliterations(phrase):
    """Iterator for all words that are part of an alliteration in given text"""
    phones_dict = {}
    phrase = removeNonWords(phrase)
    for word in phrase:
        word = word.lower()
        try:
            if (word not in stopwords or word in pronunciations):
                phones_lst = pronunciations[word][0].split()
                phones_dict[word] = phones_lst
        except:
            continue
    count = 0
    key_lst = list(phones_dict.keys())
    for cur in range(0, len(phones_dict)):
        if (cur+1 < len(phones_dict)):
            for next_index in range(cur+1, len(phones_dict)):
                if (phones_dict[key_lst[cur]][0] == phones_dict[key_lst[next_index]][0]):
                    count += 1
    return count
```

```python
alliterations("Behind every great man is a great woman, and behind every great woman is some guy staring at her behind!")
```

4. Apply this finding to entire dataset of the one-liners.txt

```python
alliteration_dict = {}
for i in range (len(humor)):
    alliteration_dict[i] = alliterations(humor[i][0])
```

### Stylistic Indicator - Antonymy

```python
nltk.download('wordnet')
import nltk
from nltk.corpus import wordnet
```

```python
def find_ant(word):
    ant = list()
    for synset in wordnet.synsets(word):
        for lemma in synset.lemmas():
            if lemma.antonyms():
                ant.append(lemma.antonyms()[0].name())
    return set(ant)
    
def num_ant(phrase):
    
    ant_dict = {}
    count = 0
    phrase = removeNonWords(phrase)
    
    for word in phrase:
        antonyms = find_ant(word)
        for antonym in antonyms:
            if antonym in phrase:
                count += 1
    return count
  
```

```python
antonym_dict = {}
for i in range (len(humor)):
    antonym_dict[i] = num_ant(humor[i][0])
```
